# Solução

Construtores !

Construtores são parâmetros que passamos às variáveis de forma que sempre em que ela é inicializada, ela apresenta esses valores

No nosso caso, queremos que todo ponto na Coordenada tenha um x e um y associado

No Python, o construtor é definido pelo método expecial *__init__*

```python
class Coordinate(object):
    """ A coordinate made up of an x and y value """
    def __init__(self, x, y):
        """ Sets the x and y values """
        self.x = x
        self.y = y
```

```python
k = Coordinate()
```

```python
k = Coordinate(3)
```

```python
k = Coordinate(3,4)
```

Além disso, em programação orientada a objetos, o método associado a classe tem o próprio objeto instaciado como um argumento. Como consequência, precisamos adicionar esse *self*


```python
z = Coordinate.__init__(z, 5,6)
```

Além das características (**atributos**), objetos associados à classes também possuem **métodos**. Métodos nada mais são do que funções específicas de uma determinada Classe.

```python
class Coordinate(object):
    """ A coordinate made up of an x and y value """
    def __init__(self, x, y):
        """ Sets the x and y values """
        self.x = x
        self.y = y
    def __str__(self):
        """ Returns a string representation of self """
        return "<" + str(self.x) + "," + str(self.y) + ">"
    def distance(self, other):
        """ Returns the euclidean distance between two points """
        x_diff_sq = (self.x-other.x)**2
        y_diff_sq = (self.y-other.y)**2
        return (x_diff_sq + y_diff_sq)**0.5
```

```python
k = Coordinate(3,4)
print (k)
```

```python
z = Coordinate(5,6)
print(z)
```

```python
z.distance(k)
```
